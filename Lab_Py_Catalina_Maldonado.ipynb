```python
#Número primos
def es_primo(n):
  if n < 2:
    return False
  for i in range (2, int(n**0.5)+1):
    if n % i == 0:
      return False
  return True
```

```python
es_primo(194950)
```

Secc de Fibonacci

```python
def fibonacci (n):
  fib_seq = [0,1]
  for i in range (2,n):
    fib_seq.append(fib_seq[-1]+ fib_seq[-2])
  return fib_seq
```

```python
fibonacci(15)
```

```python
#Contar palabras de un texto
def contar_palabras(texto):
  palabras = texto.split()
  conteo = {}
  for palabra in palabras:
    conteo[palabra] = conteo.get(palabra,0) + 1
  return conteo
```

```python
texto = ("Había una vez un perrito que se llamaba pegamento, se cayó y se pego")
contar_palabras(texto)
```

```python
es_palindromo("radar")
```

```python
#Factorial
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n-1)
```

```python
factorial (85)
```

```python
#Suma de los digitos
def suma_digitos(n):
  return sum(int(digito) for digito in str(n))
```

```python
suma_digitos(1234)
```

```python
#Encontrar el mayor de 3 números
def mayor_de_3(a,b,c):
  return max(a,b,c)
```

```python
mayor_de_3(55,33,189)
```
